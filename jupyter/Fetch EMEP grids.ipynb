```python
import requests
import zipfile
import os

def download_file(url, save_path):
    response = requests.get(url)
    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            file.write(response.content)
        print(f"File successfully downloaded and saved to {save_path}")
    else:
        print(f"Failed to download file for {save_path}. HTTP Status code: {response.status_code}")

def unzip_file(zip_path, extract_to):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
    print(f"File successfully unzipped to {extract_to}")

def generate_url(iso2_code):
    base_url = "https://webdab01.umweltbundesamt.at/download/grid_definition/EMEP_grid_01deg_shp_"
    return f"{base_url}{iso2_code}.zip?cgiproxy_skip=1"

countries = [
    "AL", "AM", "AT", "AZ", "BY", "BE", "BA", "BG", "HR", "CY", "CZ", "DK", "EE", 
    "FI", "FR", "GE", "DE", "GR", "HU", "IS", "IE", "IT", "KZ", "KG", "LV", "LI", 
    "LT", "LU", "MK", "MT", "MD", "MC", "ME", "NL", "NO", "PL", "PT", "RO", "RU", 
    "RS", "SK", "SI", "ES", "SE", "CH", "TR", "UA", "GB"
]

save_directory = 'Q:/Delivery/GISdata/CLRTAP/data/EMEP_grids/'

for country in countries:
    url = generate_url(country)
    save_path = os.path.join(save_directory, f"EMEP_grid_01deg_shp_{country}.zip")
    download_file(url, save_path)
    unzip_file(save_path, save_directory)
    os.remove(save_path)  # Optionally, remove the zip file after extraction
```
